```python
import matplotlib.pyplot as plt
import numpy as np


t = np.linspace(0, 5, 1000) 
f = 1  
damped_sine_wave = np.exp(-t) * np.sin(2 * np.pi * f * t)
upper_envelope = np.exp(-t)
lower_envelope = -np.exp(-t)
plt.figure(figsize=(10, 5))
plt.plot(t, damped_sine_wave, color='black', label=r'$e^{-t} \sin(2\pi f t)$')
plt.plot(t, upper_envelope, 'r--', label='Upper Envelope: $e^{-t}$')
plt.plot(t, lower_envelope, 'r--', label='Lower Envelope: $-e^{-t}$')
plt.title(r'$e^{-t} \sin(2\pi f t)$ with Envelopes')
plt.xlabel('t')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x and y arrays
x = np.linspace(-2, 2, 50)
y = np.exp(x)

# Modify y values: Invert every second element
for i in range(len(y)):
    if (i + 1) % 2 == 0:  # Using (i + 1) to count from 1
        y[i] = -y[i]

print(y)

# Plot the results
plt.figure(figsize=(5,5), layout="constrained")
plt.stem(-x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Stem Plot with Modified y Values')
plt.show()
```
